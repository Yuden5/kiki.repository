```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df_cleaned['price'], bins=20)
plt.title('Distribution of Airbnb Prices in Bangkok')
plt.show()
```

```python
import numpy as np

# Add a log-price column to your dataframe (avoid errors on price <=0)
df_cleaned['log_price'] = np.log1p(df_cleaned['price'])  # log1p handles 0 prices safely
#The base is e(Euler's number  2.71828)
sns.histplot(df_cleaned['log_price'], bins=13, kde=True)
plt.xlabel('Log of Price')
plt.title('Distribution of Log Prices')
plt.show()
```

```python
sns.boxplot(x='room_type', y='price', data=df_cleaned)
plt.title('Price by Room Type')
plt.show()
```

```python
import folium

# Create a map centered in Bangkok
bangkok_map = folium.Map(location=[13.7563, 100.5018], zoom_start=12)

# Plot sample of points to avoid overloading the map
sample_df = df_cleaned.sample(500, random_state=1)

for idx, row in sample_df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=2,
        color='blue',
        fill=True,
        fill_opacity=0.5
    ).add_to(bangkok_map)

bangkok_map.save('bangkok_airbnb_map.html')
```

```python
avg_price_neighbourhood = df_cleaned.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)
avg_price_neighbourhood
```

```python
max_listings_neighbourhood = df_cleaned.groupby('neighbourhood')['calculated_host_listings_count'].max().sort_values(ascending=False)
max_listings_neighbourhood
```

```python
sns.scatterplot(x='availability_365', y='price', data=df_cleaned, hue='room_type')
plt.title('Availability vs. Price')
plt.show()
```

```python
grouped_neigh_hostid = df_cleaned.groupby(['neighbourhood', 'host_id'])['number_of_reviews'].sum().reset_index()
grouped_neigh_hostid.head()
```

```python
top_hosts = grouped_neigh_hostid.sort_values(['neighbourhood', 'number_of_reviews'], ascending=False)
top_hosts.head()
```
